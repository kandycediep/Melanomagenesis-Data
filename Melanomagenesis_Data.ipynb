```python
import matplotlib.pyplot as plt
import pandas as pd
from IPython.display import display, clear_output

from google.colab import drive
drive.mount('/content/drive')
```

```python
file_path = "/content/drive/MyDrive/monotherapy.csv"
df = pd.read_csv(file_path)
df.head()

x = df['Patient']
y1 = df['PFS (weeks)']
y2 = df['OS (weeks)']

plt.plot(x, y1, marker='o', linestyle='-', color='b', label='PFS')
plt.plot(x, y2, marker='o', linestyle='-', color='r', label='OS')

plt.xlabel('Patient')
plt.ylabel('Weeks')
plt.title('Progression-Free Survival (PFS) and Overall Survival (OS) of Dabrafenib')

plt.legend(loc='upper right')

plt.show()
```

```python
categories = ['RES', 'PR', 'SD', 'No RES']
values = [1, 8, 3, 1]

plt.bar(categories, values, color='green')

plt.xlabel('RECIST Response Category')
plt.ylabel('# of patients')
plt.title('Responses to Dabrafenib')

plt.show()
```

```python
file_path2 = '/content/drive/MyDrive/combotherapy.csv'
df2 = pd.read_csv(file_path2)

fig, ax = plt.subplots(figsize=(7.5, 6))

bar_width = 0.35
index = df2.index

bar1 = ax.bar(index, df2['BRAFi'], width=bar_width, label='BRAFi (dabrafenib)')
bar2 = ax.bar(index + bar_width, df2['BRAFi + MEKi'], width=bar_width, label='BRAFi + MEKi (dabrafenib + trametinib)')

ax.set_ylabel('Weeks')
ax.set_title('BRAFi vs. BRAFi + MEKi')
ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels(df2['Category'])
ax.legend()

plt.show()
```
